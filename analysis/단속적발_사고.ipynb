```python
from analysis.querySql import query_sql
query = """SELECT
  c.구,
  c.적발건수,
  a.사고건수
FROM (
  SELECT COUNT(id) AS 적발건수, 구
  FROM crackdown
  GROUP BY 구
) c
LEFT JOIN (
  SELECT COUNT(id) AS 사고건수, 구
  FROM accident_detail
  GROUP BY 구
) a ON c.구 = a.구
ORDER BY c.적발건수 DESC;
"""
df3 = query_sql(query)
df3
```

```python
from statsmodels.formula.api import ols,glm

Rformula ="사고건수 ~ 적발건수"
regression_result = ols(Rformula, data = df3).fit()
regression_result.summary()
# 적발과 사고건수는 비례함
```
