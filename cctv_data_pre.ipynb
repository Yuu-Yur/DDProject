```python
import pandas as pd

```

```python
# 24년 자료라 크게 의미가 없음
df1 = pd.read_excel('./data/서울시 자치구 (목적별) CCTV 설치현황_241231.xlsx', header = 3)
df1.rename(columns={'소 계':'범죄예방및수사계', '어린이\n보호구역':'어린이보호구역', '공원·놀이터':'공원,놀이터', '쓰레기\n무단투기':'쓰레기무단투기','Unnamed: 8':'시설안전,화재예방','Unnamed: 9':'교통단속','Unnamed: 10':'교통정보수집,분석','Unnamed: 11':'기타목적','Unnamed: 2':'총계','Unnamed: 1':'시군구'}, inplace = True)
df1.drop(df1.columns[0],axis=1,inplace = True)
df1.fillna(value = 0, inplace = True)
df1[df1.columns[1:]] = df1.iloc[:,1:].astype(int)
df1[:26]
```

```python
df2 = pd.read_excel('./data/서울시 자치구 (범죄예방 수사용) CCTV 설치현황_241231.xlsx', header = 2)
df2.drop(columns=list(df2.columns[:2]) + list(df2.columns[3:11]) + list(df2.columns[12:]),axis=1,inplace = True)
df2.rename(columns={'2023년':'범죄예방수사용','구분':'시군구'}, inplace = True)
df2 = df2[:26]
df2['범죄예방수사용'] = df2['범죄예방수사용'].astype(int)
df2
```

```python
df3 = pd.read_excel('./data/서울시 자치구 (연도별) CCTV 설치현황_241231.xlsx', header = 2)
df3['2023년계'] = df3[df3.columns[4:13]].sum(axis = 1)
df3.drop(columns=list(df3.columns[:2]) + list(df3.columns[3:14]),axis=1,inplace = True)
df3.rename(columns={'2023년계':'2023년총CCTV','구분':'시군구'}, inplace = True)
df3['시군구'] = df3['시군구'].str.replace(' ','')
df3['2023년총CCTV'] = df3['2023년총CCTV'].astype(int)
df3[:26]
```

```python
df_concat = df2.merge(df3, on='시군구')
df_concat
```
